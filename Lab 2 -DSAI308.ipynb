# Linear Regression with PyTorch

```python
# Init some data (X, y)
x = torch.tensor([1,2,4,5,6,7,8,9], dtype=float)
y = 20 * x + 300

# Initialize Weight & Bias
w = torch.tensor([1.0], dtype=float,requires_grad=True)
b = torch.tensor([1.0], dtype=float, requires_grad=True)

# Suggest Learning Rate
lr = 0.003

# Iterate
for i in range(1000):
    y_pred = w * x + b # Forward pass
    loss = ((y_pred - y) ** 2).sum() # Loss calculation
    loss.backward() # Backward pass

    # Freeze gradients (do not calc derivatives on W,b when updating weights)
    with torch.no_grad():
        w -= lr * w.grad
        b -= lr * b.grad

    # Reset W, B derivatives
    w.grad.zero_()
    b.grad.zero_()

# Print best values for W & B
print(w.item(), b.item())
```

```python
# Get Y prediction
y_pred = (w * x + b).detach().numpy().astype(np.float16) # Detach
y_pred
```

# Loss Function

```python
# Init some data
x = torch.tensor([1,2,4,5,6,7,8,9], dtype=float)
y = 20 * x + 300

# Init W & b
w = torch.tensor([1.0], dtype=float,requires_grad=True)
b = torch.tensor([1.0], dtype=float, requires_grad=True)

# Set MSE Loss Function
mse = torch.nn.MSELoss()

# Optimizer (Gradient Descent variants)
optimizer = torch.optim.SGD([w,b], lr=0.01)

# Training
for i in range(1000):
    y_pred = w * x + b # Forward pass
    loss = mse(y, y_pred) # Loss
    loss.backward() # Backpropagation
    optimizer.step() # Update weights
    optimizer.zero_grad() # Clear gradients


print(w.item(), b.item())
```

# Neural Networks

```python
import torch.nn as nn
```

```python
x = torch.tensor([1,2,4,5,6,7,8,9], dtype=torch.float32).view(-1,1)
y = 20 * x + 300
```

```python
# Define a class that inherits from nn.Module
class LinearRegression(nn.Module):
    def __init__(self):
        super(LinearRegression, self).__init__()

        # Dense layer
        self.linear = nn.Linear(1, 1)

    # Forward pass logic
    def forward(self, x):
        # Pass input (x) to layer
        return self.linear(x)
```

```python
epochs = 2000
lr = 0.01 # Learning Rate
model = LinearRegression()
mse = nn.MSELoss() # Loss
optimizer = torch.optim.SGD(model.parameters(), lr)
```

```python
# Training
for i in range(epochs+1):
    y_pred = model(x) # Forward (get output)
    loss = mse(y, y_pred) # Loss
    loss.backward() # Backprop
    optimizer.step() # Update weights
    optimizer.zero_grad() # Clear gradients
    if i % 100 == 0:
        print('epoch {}, loss {}'.format(i, round(loss.item(),3)))
```

```python
class MyDataset(Dataset):

  # Read & process dataset
    def __init__(self):
        df = pd.read_csv("/content/IRIS.csv")
        self.x = torch.from_numpy(df.iloc[:, :-1].values.astype(np.float32))
        self.y = torch.from_numpy(pd.get_dummies(df.iloc[:,-1]).values.argmax(axis=1)).long()
        self.n_samples = df.shape[0]

    def __getitem__(self, index):
        return self.x[index], self.y[index]

    def __len__(self):
        return self.n_samples
```

# MNIST Dataset

```python
import torch
import torch.nn as nn
import torch.nn.functional as F
from torchvision import datasets
from torchvision.transforms import transforms
from torch.utils.data import DataLoader, Dataset
```

<img src="https://user-images.githubusercontent.com/50156227/204368149-10d739d8-935c-402c-97af-336255ec76d9.gif">

```python
input_size = 28 * 28 # image shape
n_classes = 10
hidden_size = 128
epochs = 10
batch_size = 128
lr = 0.001
```

```python
# Transformation Unit (Class) to reshape MNIST images (28 x 28) -> 784
class Reshape():
    def __init__(self, shape):
        self.shape = shape
    def __call__(self, arr):
        return arr.reshape(self.shape)
```

```python
# Transformation Composition (think of it as a pipeline\chain of processes)
composed = transforms.Compose([transforms.ToTensor(), Reshape(input_size)])
```

```python
# Load datasets & apply transformation pipeline
train_dataset = datasets.MNIST(root="/", transform=composed, download=True)
test_dataset = datasets.MNIST(root="/", transform=composed, download=False, train=False)
```

```python
train_dataset
```

```python
# Init dataset loaders
train_loader = DataLoader(train_dataset, batch_size=batch_size, shuffle=True)
test_loader = DataLoader(test_dataset, batch_size=batch_size, shuffle=True)
```

```python
# Define Neural Network
class NeuralNet(nn.Module):
    def __init__(self, in_dim, hidden_dim, n_classes):
        super(NeuralNet, self).__init__()
        self.lin = nn.Linear(in_dim, hidden_dim)
        self.lin2 = nn.Linear(hidden_dim, n_classes)

    def forward(self, x):
        out = self.lin(x)
        out = F.relu(out) # activation function
        out = self.lin2(out)
        return out
```

```python
model = NeuralNet(input_size, hidden_size, n_classes)
```

```python
criterion = nn.CrossEntropyLoss() # Loss function
optimizer = torch.optim.Adam(model.parameters(), lr=lr)
```

```python
# Epochs
for epoch in range(epochs):
    # Keep track of samples & outputs to calc accuracy
    n_correct = 0
    n_samples = 0
    train_loss = 0

    # Forward pass through batches
    for i, (images, labels) in enumerate(train_loader):
        y_pred = model(images)
        loss = criterion(y_pred, labels)
        train_loss += loss.item() # Keep track of losses over epochs
        loss.backward()

        optimizer.step()
        optimizer.zero_grad()

        # Count total samples & correct ones for current batch
        with torch.no_grad():
            logit_outputs = torch.max(y_pred, axis=1).indices
            n_correct += (labels == logit_outputs).sum()
            n_samples += labels.size(0)

    # Calculate accuracy & loss after epoch
    with torch.no_grad():
        # Training Acc & avg loss per batch
        train_acc = (n_correct / n_samples)
        train_loss = train_loss / (i+1)

        # Re-use counters for Validation Accuracy & Loss
        n_correct = 0
        n_samples = 0
        val_loss = 0

        # Forward pass through test/validation dataloader
        for i, (images, labels) in enumerate(test_loader):
            predictions = model(images) # Inference
            val_loss += criterion(predictions, labels) # Accumulate val loss

            # Count samples & correct outputs
            logit_outputs = torch.max(predictions, axis=1).indices
            n_correct += (labels == logit_outputs).sum()
            n_samples += labels.size(0)

        # Acc = correct answers / total samples
        val_acc = (n_correct / n_samples)
        val_loss = val_loss / (i+1)

        msg = f"Epoch {epoch+1}/{epochs}: loss: {train_loss:.4f} - acc: {train_acc.item():.4f} - val_loss: {val_loss.item():.4f} - val_acc: {val_acc.item():.4f}"
        print(msg)
```
